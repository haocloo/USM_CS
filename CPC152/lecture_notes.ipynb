Matrix operations
===

```python
import numpy as np

v = np.array([1,2])
u = np.array([3,4])
```

```python
print(v + u)
print(v * u)
```

```python
print("dot = 1x3 + 2x4:")
print(v.dot(u))
```

```python
print("double bar v = sqrt(1^2 + 2^2) :")
print(np.linalg.norm(v))
```

```python
print("unit vector")
print(v / np.linalg.norm(v))
```

```python
print('angle between vectors in degrees:')
norms = np.linalg.norm(v) * np.linalg.norm(u)
print(np.rad2deg(np.arccos(v.dot(u) / norms)))
```

```python
a = np.matrix([[1,2],[3,4]])
print(a)
print(a.T)
```

```python
print(np.eye(3))
```

```python
print("determinant:")
print(np.linalg.det(a))
```

```python
# determinant cannot be zero, must be square matrix (2x2, 3x3, etc)
print("inverse:")
print(np.linalg.inv(a))
print("check:")
print(a.dot(np.linalg.inv(a)))
```

## Box plot

```python
value1=[82,76,24,40,67,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
value2=[62,5,91,25,36,32,96,95,3,90,95,32,27,55,100,15,71,11,37,21]
value3=[23,89,12,78,72,89,25,69,68,86,19,49,15,16,16,75,65,31,25,52]
value4=[59,73,70,16,81,61,88,98,10,87,29,72,16,23,72,88,78,99,75,30]
 
box_plot_data=[value1,value2,value3,value4]

box = plt.boxplot(box_plot_data,patch_artist=True,labels=['course1','course2','course3','course4'])
# horizontal box plot
# box = plt.boxplot(box_plot_data,vert=0,patch_artist=True,labels=['course1','course2','course3','course4'])

colors = ['cyan', 'lightblue', 'lightgreen', 'tan']
for patch, color in zip(box['boxes'], colors): # zip() function is used to iterate over two or more iterables at the same time.
    patch.set_facecolor(color)
    
plt.show()
```
