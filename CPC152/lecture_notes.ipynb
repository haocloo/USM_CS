```python
weight1=[57,58.2,58.6,59.6,59.8,60.2,60.5,60.6,60.7,61.3,61.3,61.4,61.8,61.9,62.3]
height1=[100.7,195.6,94.3,127.1,111.7,159.7,135,149.9,124.3,112.9,176.7,110.2,123.9,161.9,107.8]
 
weight2=[62.9,63,63.1,63.2,63.3,63.4,63.4,63.4,63.5,63.6,63.7,64.1,64.3,64.3,64.7,64.8,65]
height2=[151.8,156.3,136.1,124.2,156.3,130,181.2,255.9,163.1,123.1,119.5,179.9,114.8,174.1,108.8,105.4,141.4]
 
weight3=[69.2,69.2,69.4,69.7,70,70.3,70.8,71,71.1,71.7,71.9,72.4,73,73.1,76.2]
height3=[166.8,172.9,193.8,137.4,162.4,137.1,169.1,237.1,189.1,179.3,174.8,213.3,198,191.1,220.6]

weight=np.concatenate((weight1,weight2,weight3))
height=np.concatenate((height1,height2,height3))
 
#Corlor_array with first 15 blue color, second 15 green color and last 15 red color is created.
color_array = ['b'] * len(weight1) + ['g'] * len(weight2) + ['r'] * len(weight3) #blue, green, red for 3 groups
 
#All the above three arguments along with the marker=’*’ is passed to Scatter function.
plt.scatter(weight, height, marker='*', c=color_array)
 
plt.xlabel('weight', fontsize=10)
plt.ylabel('height', fontsize=10)
plt.title('grouped scatter plot - height vs weight',fontsize=10)
plt.show()
```

## Box plot

```python
value1=[82,76,24,40,67,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
value2=[62,5,91,25,36,32,96,95,3,90,95,32,27,55,100,15,71,11,37,21]
value3=[23,89,12,78,72,89,25,69,68,86,19,49,15,16,16,75,65,31,25,52]
value4=[59,73,70,16,81,61,88,98,10,87,29,72,16,23,72,88,78,99,75,30]
 
box_plot_data=[value1,value2,value3,value4]

box = plt.boxplot(box_plot_data,patch_artist=True,labels=['course1','course2','course3','course4'])
# horizontal box plot
# box = plt.boxplot(box_plot_data,vert=0,patch_artist=True,labels=['course1','course2','course3','course4'])

colors = ['cyan', 'lightblue', 'lightgreen', 'tan']
for patch, color in zip(box['boxes'], colors): # zip() function is used to iterate over two or more iterables at the same time.
    patch.set_facecolor(color)
    
plt.show()
```

Read the `unemployment1948.csv` file into a Pandas data frame being careful to set an appropriate index. This is US unemployment rate for every month, and the annual rate, from 1948 to 2013.

Plot the Jan, Apr, Jul, and Oct unemployment rates along with the annual unemployment rates in one figure. Consult the [Plot documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) to use the `color=`, `linestyle=`, `linewidth=` and `marker=` keyword arguments to make the plot.
